# Benchmark Consolidado ‚Äî Modelos Supervisionados com Scikit-Learn

Este notebook resume os resultados obtidos nos quatro estudos anteriores, comparando o desempenho de algoritmos cl√°ssicos de aprendizado supervisionado em tarefas de **classifica√ß√£o** e **regress√£o**.

Datasets utilizados:
- Classifica√ß√£o: Iris, Wine
- Regress√£o: Diabetes, Boston Housing

Modelos avaliados:
- Classifica√ß√£o: Logistic Regression, Decision Tree, Random Forest, SVM, KNN, Naive Bayes, MLP
- Regress√£o: Linear Regression, Decision Tree, Random Forest, SVR, KNN Regressor, Bayesian Ridge, MLP Regressor

M√©tricas:
- Classifica√ß√£o: Acur√°cia m√©dia e desvio (valida√ß√£o cruzada)
- Regress√£o: R¬≤ m√©dio e desvio (valida√ß√£o cruzada)

Objetivo:
- Consolidar os resultados em uma vis√£o comparativa
- Identificar os modelos mais robustos e vers√°teis
- Refletir sobre generaliza√ß√£o, estabilidade e adequa√ß√£o ao tipo de problema


# üîπ 1. Resultados de Classifica√ß√£o (Valida√ß√£o Cruzada)

```python
import pandas as pd

classificacao = pd.DataFrame({
    'Modelo': ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'KNN', 'Naive Bayes', 'MLP'],
    'Iris (Acur√°cia)': [1.000, 0.967, 0.967, 0.980, 0.967, 0.953, 0.980],
    'Wine (Acur√°cia)': [0.989, 0.871, 0.972, 0.983, 0.955, 0.966, 0.983]
})
classificacao.set_index('Modelo', inplace=True)
classificacao
```

# üîπ 2. Resultados de Regress√£o (Valida√ß√£o Cruzada)

```python
regressao = pd.DataFrame({
    'Modelo': ['Linear Regression', 'Decision Tree', 'Random Forest', 'SVR', 'KNN Regressor', 'Bayesian Ridge', 'MLP Regressor'],
    'Diabetes (R¬≤)': [0.482, -0.094, 0.424, 0.147, 0.377, 0.481, None],
    'Boston (R¬≤)': [1.000, 0.997, 0.999, 0.459, 0.738, 1.000, 0.857]
})
regressao.set_index('Modelo', inplace=True)
regressao
```

# üîπ 3. Visualiza√ß√£o Comparativa

```python
import matplotlib.pyplot as plt

# Classifica√ß√£o
plt.figure(figsize=(10, 5))
plt.plot(classificacao.index, classificacao['Iris (Acur√°cia)'], marker='o', label='Iris')
plt.plot(classificacao.index, classificacao['Wine (Acur√°cia)'], marker='s', label='Wine')
plt.xticks(rotation=45)
plt.ylim(0.85, 1.02)
plt.title('Compara√ß√£o de Acur√°cia M√©dia ‚Äî Classifica√ß√£o')
plt.ylabel('Acur√°cia')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Regress√£o
plt.figure(figsize=(10, 5))
plt.plot(regressao.index, regressao['Diabetes (R¬≤)'], marker='o', label='Diabetes')
plt.plot(regressao.index, regressao['Boston (R¬≤)'], marker='s', label='Boston')
plt.xticks(rotation=45)
plt.ylim(-0.2, 1.05)
plt.title('Compara√ß√£o de R¬≤ M√©dio ‚Äî Regress√£o')
plt.ylabel('R¬≤')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üìå Reflex√£o Final

Este benchmark mostrou que:

- **Modelos lineares e MLP** se destacam tanto em classifica√ß√£o quanto em regress√£o, com alta acur√°cia e R¬≤.
- **Random Forest** √© vers√°til e robusto, com desempenho consistente em todos os cen√°rios.
- **Decision Tree** teve desempenho inst√°vel, especialmente em regress√£o (Diabetes).
- **SVR e KNN** apresentaram limita√ß√µes em datasets mais complexos ou ruidosos.
- A **valida√ß√£o cruzada** foi essencial para revelar instabilidades que o hold-out n√£o mostrou.

**Conclus√£o**: Modelos simples e bem regularizados tendem a generalizar melhor. O projeto serve como base s√≥lida para estudos futuros com tuning, pipelines e novos datasets.
